```python
import pandas as pd

df = pd.read_csv("Fire-Incidents.csv")
df.head()
```

```python
neighborhoods= df["neighborhood"].value_counts().sort_values(ascending=True)
for neighborhood, num in neighborhoods.items():
    print(neighborhood, num)
```

```python
import matplotlib.pyplot as plt

neighborhoods= df["neighborhood"].value_counts().sort_values(ascending=True)

plt.figure(figsize=(10,4))
neighborhoods.plot(kind="bar")
plt.xticks(fontsize=6)
plt.show()
```

Mount Oliver Borough would be considered the best neighborhood in Pittsburgh due to it having the least amount of fire incidents; therefore, being the safest.
